#### Question 1: What is the northernmost airport in the United States?

```python
import pandas as pd
a_d = pd.read_csv("airports.csv")
```

Check what the file looks like:

```python
a_d.head()
```

Since latitude represents the north/south orientation of a point, sorting for the highest latitude within the US should give us the northernmost airport.

```python
a_d.sort_values('lat', ascending=False).head()
```

Need to remove the tzone with the null value and confirm the new record order.

```python
a_d[a_d.tzone.notnull()].sort_values('lat', ascending=False).head()
```

isolate the record with the highest lat

```python
a_d[(a_d.lat == (a_d[a_d.tzone.notnull()].lat.max()))]
```

#### Question 2: What is the easternmost airport in the United States? 

Sorting for the highest longitude should identify the easternmost airport. 

```python
a_d.sort_values('lon', ascending=False).head()
```

Anchorage Alakska appears to be the answer, however, this is apparently debatable because it's actually beyond the prime meridian. So we'll focus on the contiguous US, which means we have to find lon with negative values:

```python
a_d[(a_d.lon == (a_d[a_d.tzone.notnull() & (a_d.lon <= 0)].lon.max()))]
```

#### Question 3: On February 12th, 2013, which New York area airport had the windiest weather? 

Need to grab the weather file and spot check the file contents:

```python
w = pd.read_csv("weather.csv")
```

```python
w.head()
```

Filtering for Feb 12, 2013:

```python
w[(w.year == 2013) & (w.month == 2) & (w.day == 12)].head()
```

Add grouping to find the windiest weather by airport:

```python
w[(w.year == 2013) & (w.month == 2) & (w.day == 12)].groupby('origin').wind_speed.max()
```

Since hurricane force winds are less than 200 mph, the 1048 wind speed is probably just an error:

```python
w[(w.year == 2013) & (w.month == 2) & (w.day == 12) & (w.wind_speed <= 200)].groupby('origin').wind_speed.max()
```

#### <font color= green> __LGA had the windiest weather on Feb 12, 2013 at 23 miles per hour.__ </font>
